```
#@title Fine Tuning Parameters
import keras
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


Hidden_Layers = 4 #@param {type:"slider", min:1, max:4, step:1}
Optimizer = "rmsprop" #@param ["adam", "rmsprop"]
Loss = 'binary_crossentropy' #@param ["categorical_crossentropy", "binary_crossentropy"]
DropOut = 0.5 #@param {type: "slider", min:0.1, max :0.5, step: 0.1}
Activation = "sigmoid" #@param ["sigmoid", "softmax"]
File_Name = 'Cancer Classification' #@param {type:"string"}
Batch_Size = 40 #@param {type: "slider", min:10, max :40, step: 1}
Epochs = 500 #@param {type: "slider", min:100, max:500, step: 100}

#Getting x and y values
dataset = pd.read_csv('breast_cancer.csv')
X = dataset.iloc[:, 2:31].values
y = dataset.iloc[:, 1].values

#Encoding the malignant and benign
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()

y = labelencoder.fit_transform(y)

#Splitting training data and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

#Feature Scaling
from sklearn.preprocessing import StandardScaler
fs = StandardScaler()
X_train = fs.fit_transform(X_train)
X_test = fs.transform(X_test)

from keras.models import Sequential
from keras.layers import Dense, Dropout

classifier = Sequential()
#First layer
classifier.add(Dense(activation = "relu", input_dim = 29, units = 16, kernel_initializer = "uniform"))
classifier.add(Dropout(DropOut))
#Hidden Layers
for i in range(Hidden_Layers):
  classifier.add(Dense(activation = "relu", units = 16, kernel_initializer = "uniform"))
  classifier.add(Dropout(DropOut))
#Output layer
classifier.add(Dense(activation = Activation, units = 1, kernel_initializer = "uniform"))

#Compiling the ANN
classifier.compile(optimizer = Optimizer, loss = Loss, metrics = ['accuracy'])

classifier.fit(X_train, y_train, batch_size = Batch_Size, epochs = Epochs)


# Part 3 - Making predictions and evaluating the model


# Predicting the test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)


# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

print(y_pred)
print(cm)
```
